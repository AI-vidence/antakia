# 🚀 Exemple d'utilisation optimisée d'Antakia - Dataset Wages

Ce notebook montre comment utiliser Antakia avec les optimisations de performance pour un démarrage rapide.


````python
# 🔧 ÉTAPE 1: Configurer les performances AVANT d'importer Antakia
from antakia.performance_config import apply_fast_config
apply_fast_config()  # Configuration rapide: PCA + LIME + limite de points
````

````python
# 📊 ÉTAPE 2: Charger les données
import pandas as pd
from antakia.utils.examples import fetch_dataset

df = fetch_dataset('wages')
X = df.iloc[:,0:3] # le dataset
y = df.loc[:,'salaire'] # la variable cible

print(f"Dataset shape: {df.shape}")
print(f"⚡ Prêt pour Antakia optimisé!")
df.head()
````

````python
# 🤖 ÉTAPE 3: Entraîner le modèle (rapide avec moins d'estimateurs)
from sklearn.ensemble import GradientBoostingRegressor
import time

model = GradientBoostingRegressor(random_state=9, n_estimators=50)  # Réduit pour rapidité
start = time.time()
model.fit(X, y)
train_time = time.time() - start

print(f"⚡ Modèle entraîné en {train_time:.1f}s. Score: {model.score(X, y):.3f}")
````

````python
# 🚀 ÉTAPE 4: Utiliser Antakia OPTIMISÉ (démarrage ultra-rapide!)
from antakia.antakia import AntakIA

start = time.time()
atk = AntakIA(X, y, model)
creation_time = time.time() - start

print(f"🎉 AntakIA créé en {creation_time:.1f}s (au lieu de 2-10 minutes!)")
print("📝 Les explications SHAP/LIME seront calculées à la demande si nécessaire")

# Lancer l'interface
atk.start_gui()
````

## 💡 Conseils pour optimiser davantage

### Si c'est encore trop lent :
```python
from antakia.performance_config import apply_minimal_config
apply_minimal_config()  # Configuration ultra-rapide
```

### Autres configurations disponibles :
```python 
from antakia import performance_config as perf
perf.apply_medium_config()    # Équilibre performance/qualité
perf.apply_quality_config()   # Qualité maximale (lent)
```

### Pour de très gros datasets :
```python
# Sous-échantilloner les données pour les tests
if len(df) > 5000:
    df = df.sample(n=5000, random_state=42)
```
